### ▶ Unit 1 Capstone — Compound drainage catchment design

```python
# ═══════════════════════════════════════════════════════════
# CE541E08 — Unit 1 Capstone Assignment
# PROBLEM: Compound catchment → outfall pipe design
# ═══════════════════════════════════════════════════════════
import math

# ── CATCHMENT DATA ───────────────────────────────────────────
areas_ha = [8.5, 12.3, 5.7]
c_values = [0.70, 0.55, 0.80]
intensity_mmhr = 48.0

# ── PIPE DATA ────────────────────────────────────────────────
D_mm = 600; n = 0.013; S = 0.0015; L = 280.0

# ════════════════════════════════════════════════════════════
# STEP 1: RATIONAL METHOD
# ════════════════════════════════════════════════════════════
i_ms        = intensity_mmhr / 1000 / 3600
total_ha    = sum(areas_ha)
total_m2    = total_ha * 10000
C_weighted  = sum(c*a for c,a in zip(c_values,areas_ha)) / total_ha
Q_peak      = C_weighted * i_ms * total_m2

print("STEP 1 — PEAK DISCHARGE")
print("─"*50)
print(f"  Total area   : {total_ha:.1f} ha  ({total_m2:,.0f} m²)")
print(f"  Weighted C   : {C_weighted:.3f}")
print(f"  Q_peak       : {Q_peak:.4f} m³/s  ({Q_peak*1000:.2f} L/s)")

# ════════════════════════════════════════════════════════════
# STEP 2: PIPE CAPACITY
# ════════════════════════════════════════════════════════════
D    = D_mm/1000
A_p  = math.pi*(D/2)**2
R_p  = A_p/(math.pi*D)
V_p  = (1/n)*R_p**(2/3)*S**0.5
Q_cap = V_p*A_p

print("STEP 2 — PIPE CAPACITY")
print("─"*50)
print(f"  Pipe {D_mm}mm, n={n}, S=1:{int(1/S)}")
print(f"  Velocity     : {V_p:.3f} m/s")
print(f"  Q_capacity   : {Q_cap:.4f} m³/s  ({Q_cap*1000:.2f} L/s)")

# ════════════════════════════════════════════════════════════
# STEP 3: DESIGN CHECKS
# ════════════════════════════════════════════════════════════
cap_ratio = Q_peak/Q_cap
vel_ok    = 0.6<=V_p<=3.0
cap_ok    = Q_cap>=Q_peak
Re        = V_p*D/1.007e-6

print("STEP 3 — DESIGN CHECKS")
print("─"*50)
print(f"  Q_peak/Q_cap : {cap_ratio:.3f}  ({cap_ratio*100:.1f}% full)")
print(f"  Capacity OK? : {'YES ✓' if cap_ok else 'NO ✗'}")
print(f"  Velocity OK? : {'PASS ✓' if vel_ok else 'FAIL ✗'}")
print(f"  Re           : {Re:,.0f}  Turbulent={'✓' if Re>4000 else '✗'}")
print("=" * 50)
print(f"  VERDICT: {'ACCEPTABLE ✓' if cap_ok and vel_ok else 'REVIEW REQUIRED ✗'}")
print("="*50)

# ANSWER KEY: total=26.5 ha, C_w=0.6434, Q_peak=0.02270 m³/s
# Q_cap≈0.33 m³/s — pipe operates at ~6.8% capacity (over-designed)
```
