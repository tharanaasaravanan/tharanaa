### ▶ Cell 4 — Compute cross-sectional area using π

```python
# ─────────────────────────────────────────────────────────────
# WHAT THIS CELL DOES:
# Computes the cross-sectional area of a circular pipe.
# Formula: A = π/4 × D²
# We import the math module to get a precise value of π.
# "import" means: bring in a toolkit of extra functions.
# ─────────────────────────────────────────────────────────────

import math             # this loads Python's maths toolkit

# Pipe parameters
diameter = 0.3          # metres (300 mm pipe)
radius   = diameter / 2 # radius = D/2

# Cross-sectional area: A = π r²  (same as π/4 × D²)
area = math.pi * radius ** 2

# Wetted perimeter of a full circular pipe: P = π D
perimeter = math.pi * diameter

# Display results
print("─── Pipe Cross-Section Calculator ───")
print("Diameter        :", diameter,        "m")
print("Radius          :", radius,          "m")
print("Cross-sec. area :", round(area, 4),  "m²")
print("Wetted perimeter:", round(perimeter, 4), "m")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Ask the class: "Verify this by hand."
# A = π/4 × 0.3² = 3.1416/4 × 0.09 = 0.0707 m²  ✓
# P = π × 0.3 = 0.9425 m  ✓
#
# Key teaching point:
# round(area, 4) means "round to 4 decimal places"
# Without round(), Python gives: 0.07068583470577035
# Engineers always round to meaningful precision.
# ───────────────────────────────────────────────────────────────
```

## Section 3 — Manning's Equation in Python

Manning's equation gives the flow velocity in an open channel or pipe:

$$V = \frac{1}{n} \cdot R^{2/3} \cdot S^{1/2}$$

Where:
- **V** = velocity (m/s)
- **n** = Manning's roughness coefficient (dimensionless)
- **R** = hydraulic radius = Area / Wetted perimeter (m)
- **S** = longitudinal slope (m/m)

You have computed this by hand many times. Now let Python do it.

### ▶ Cell 6 — Manning's equation: velocity and discharge in a circular pipe

```python
# ─────────────────────────────────────────────────────────────
# WHAT THIS CELL DOES:
# Solves Manning's equation for a circular pipe flowing full.
# V = (1/n) × R^(2/3) × S^(0.5)
# Q = V × A
# ─────────────────────────────────────────────────────────────

import math

# ── PIPE PARAMETERS — change these to explore ───────────────
diameter = 0.3      # pipe diameter in metres
n        = 0.013    # Manning's n for concrete pipe (typical)
slope    = 0.001    # longitudinal slope (1 in 1000)
# ────────────────────────────────────────────────────────────

# Geometry (for a full circular pipe)
radius        = diameter / 2
area          = math.pi * radius ** 2
wet_perimeter = math.pi * diameter
hyd_radius    = area / wet_perimeter     # R = A / P

# Manning's equation
velocity      = (1 / n) * (hyd_radius ** (2/3)) * (slope ** 0.5)

# Discharge
discharge     = velocity * area          # Q = V × A

# ── Output ──────────────────────────────────────────────────
print("══════════════════════════════════════════")
print("    Manning's Equation — Pipe Flow Result")
print("══════════════════════════════════════════")
print(f"  Diameter            : {diameter*1000:.0f} mm")
print(f"  Manning's n         : {n}")
print(f"  Slope               : 1 in {int(1/slope)}")
print(f"  Hydraulic radius    : {hyd_radius:.4f} m")
print(f"  Flow velocity V     : {velocity:.3f} m/s")
print(f"  Discharge Q         : {discharge*1000:.2f} litres/second")
print(f"  Discharge Q         : {discharge:.5f} m³/s")
print("══════════════════════════════════════════")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# VERIFICATION:
# For D=0.3m, n=0.013, S=0.001:
#   R = D/4 = 0.075 m  (for full circular pipe)
#   V = (1/0.013) × 0.075^(2/3) × 0.001^(0.5)
#     = 76.92 × 0.1797 × 0.03162
#     ≈ 0.437 m/s
#   Q = 0.437 × 0.0707 ≈ 0.031 m³/s = 30.9 L/s
#
# Ask the class to verify this by hand (5 min).
# Then ask: "What happens to velocity if slope doubles?"
# Have them change slope = 0.002 and re-run.
# ───────────────────────────────────────────────────────────────
```

### ▶ Cell 8 — Class exercise: unit conversion for water resources

```python
# ─────────────────────────────────────────────────────────────
# CLASS EXERCISE (10 minutes)
#
# WHAT THIS CELL DOES:
# Converts between common units used in water resources.
# This is something you will do constantly when working
# with IMD rainfall data and CWC streamflow records.
# ─────────────────────────────────────────────────────────────

# A river gauging station records the following:
discharge_cumecs = 450.0    # cumecs = m³/s
rainfall_mm      = 85.3     # mm per day (IMD daily rainfall)
area_km2         = 1250.0   # catchment area in km²

# ── CONVERSIONS ─────────────────────────────────────────────

# 1. Discharge: cumecs → litres/second
discharge_lps = discharge_cumecs * 1000
print(f"Discharge: {discharge_cumecs} m³/s = {discharge_lps:.0f} litres/second")

# 2. Rainfall: mm/day → m/day
rainfall_m_day = rainfall_mm / 1000
print(f"Rainfall:  {rainfall_mm} mm/day = {rainfall_m_day} m/day")

# 3. Catchment area: km² → m²
area_m2 = area_km2 * (1000 ** 2)
print(f"Area:      {area_km2} km² = {area_m2:.2e} m²")

# 4. Rainfall volume over catchment: V = rainfall depth × area
#    (this is how total runoff volume is estimated)
rainfall_volume_m3 = rainfall_m_day * area_m2
rainfall_volume_Mm3 = rainfall_volume_m3 / 1e6   # million m³
print(f"""
If all {rainfall_mm} mm fell on {area_km2} km²:""")
print(f"  Rainfall volume = {rainfall_volume_Mm3:.2f} million m³")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# EXERCISE FOR STUDENTS:
# Change discharge_cumecs to the monsoon peak flow of the
# Cauvery river at Musiri (approx 5000 m³/s during floods).
# What is this in litres per second?
# Answer: 5,000,000 litres/second — dramatic number.
# Ask: "Why do we need Python for this? Because these
# calculations repeat for every day of a 30-year record."
# ───────────────────────────────────────────────────────────────
```

**Task:** A rectangular irrigation channel has:
- Width B = 2.0 m
- Depth of flow y = 0.8 m
- Manning's n = 0.025 (earthen channel)
- Slope S = 0.0005

Write Python code (in the cell below) to compute:
1. Cross-sectional area A = B × y
2. Wetted perimeter P = B + 2y
3. Hydraulic radius R = A / P
4. Velocity V using Manning's equation
5. Discharge Q = V × A

Format your output neatly using print() statements with clear labels and units.

### ▶ Assignment cell — write your solution here

```python
# ─────────────────────────────────────────────────────────────
# ASSIGNMENT — Rectangular channel: Manning's equation
# Replace the ??? with correct Python expressions.
# ─────────────────────────────────────────────────────────────

import math

# Given
B = 2.0       # channel width, m
y = 0.8       # flow depth, m
n = 0.025     # Manning's roughness
S = 0.0005    # slope

# Compute (replace ??? with your expressions)
A = B * y       # cross-sectional area
P = B + 2 * y       # wetted perimeter
R = A / P       # hydraulic radius
V = (1/n) * (R**(2/3)) * (S**0.5)       # Manning's velocity
Q = V * A       # discharge

# Print results
print("Rectangular Channel — Manning's Equation")
print(f"  Area A     = {round(A,4)} m²")
print(f"  Perimeter P= {round(P,4)} m")
print(f"  Hyd. radius= {round(R,4)} m")
print(f"  Velocity V = {round(V,3)} m/s")
print(f"  Discharge Q= {round(Q,4)} m³/s")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# ANSWER KEY (do not share with students before submission):
# A = B * y = 1.6 m²
# P = B + 2*y = 3.6 m
# R = A / P = 0.4444 m
# V = (1/0.025) × 0.4444^(2/3) × 0.0005^0.5
#   = 40 × 0.5771 × 0.02236 = 0.516 m/s
# Q = 0.516 × 1.6 = 0.826 m³/s
# ───────────────────────────────────────────────────────────────
```
