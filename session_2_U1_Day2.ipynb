## Section 1 — The four rules of Python syntax

| Rule | What it means |
|---|---|
| **Indentation** | Spaces at the start define code blocks — not decoration |
| **Case-sensitive** | `Area` ≠ `area` |
| **Colon** | `:` starts every new block (if, for, def) |
| **Hash `#`** | Everything after # is a comment — Python ignores it |

### ▶ Cell 2 — Block structure with a pipe check

### ▶ Cell 3 — import math: pipe geometry with exact π

````python
import math

diameter = 0.6        # 600 mm trunk main
radius   = diameter / 2
area     = math.pi * radius**2
perim    = math.pi * diameter
R        = area / perim           # hydraulic radius = D/4

print(f"Diameter         : {diameter*1000:.0f} mm")
print(f"Area             : {area:.6f} m²")
print(f"Wetted perimeter : {perim:.6f} m")
print(f"Hydraulic radius : {R:.6f} m")
print(f"D/4 check        : {diameter/4:.6f} m ✓")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Hydraulic radius of a full circular pipe = D/4.
# Proof: R = A/P = (πD²/4)/(πD) = D/4.
# This is used in Manning's equation.
# ───────────────────────────────────────────────────────────────
````

### ▶ Cell 4 — A complete, well-structured engineering script

````python
# ═══════════════════════════════════════════════════
# Script  : Circular Pipe Cross-Section Calculator
# Author  : CE541E08 Student
# Purpose : Compute geometric properties of a pipe
# ═══════════════════════════════════════════════════
import math

# ── INPUTS ──────────────────────────────────────────
diameter_mm = 450          # pipe diameter, mm
# ────────────────────────────────────────────────────

diameter = diameter_mm / 1000
area     = math.pi * (diameter/2)**2
perim    = math.pi * diameter
R        = area / perim

print(f"Diameter         : {diameter_mm} mm  ({diameter} m)")
print(f"Cross-section    : {area:.4f} m²")
print(f"Wetted perimeter : {perim:.4f} m")
print(f"Hydraulic radius : {R:.4f} m")

# ─── INSTRUCTOR NOTE ───────────────────────────────
# Target format for ALL student submissions.
# Header → Inputs with units → Calculations → Output.
# ───────────────────────────────────────────────────
````

### ▶ Assignment cell

````python
# ═══════════════════════════════════════════════════
# Script  : Rectangular Channel Cross-Section
# Author  : [Your Name]
# ═══════════════════════════════════════════════════
import math

# ── INPUTS ──────────────────────────────────────────
B = 3.5    # m
y = 1.2    # m
# ────────────────────────────────────────────────────

A = B * y
P = B + 2 * y
R = A / P
D = A / B    # hydraulic depth = A/T where T=B for rectangular

print(f"Width B        : {B} m")
print(f"Depth y        : {y} m")
print(f"Area A         : {round(A,4)} m²")
print(f"Perimeter P    : {round(P,4)} m")
print(f"Hyd. radius R  : {round(R,4)} m")
print(f"Hyd. depth D   : {round(D,4)} m")
````
